```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
import random
import time
from tqdm.notebook import tqdm
from google.colab import drive
drive.mount('/content/drive')
```

```python
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36'}
song_n = 'https://www.melon.com/song/detail.htm?songId='

df1 = pd.read_csv('/content/drive/Shareddrives/Parrot/Jinho/rock/rock저장1-15000.csv',encoding='utf-8',index_col=0)

df_lyric=pd.DataFrame(columns=['가사'])
song_lyric = []
count = 0
no_lyric = []
for i in tqdm(range(8900,15000)):
  response = requests.get(song_n + str(df1['곡 번호'].values[i]), headers=header)

  # 2초 내에서 랜덤으로 쉬어주기
  time.sleep(2.5)
  
  html = response.text
  soup = BeautifulSoup(html, 'html.parser')

  lyric = soup.find('div',{'class':'lyric'})
  
  try:
    song_lyric.append(lyric.text)
  except AttributeError:
    song_lyric.append('가사 확인 불가')
    no_lyric.append(i)
  count += 1

  if count % 100 == 0:
    #중간저장
    new = pd.DataFrame(np.array([song_lyric]).T, columns=['가사'])
    df_lyric = pd.concat([df_lyric, new], ignore_index=True)
    df_lyric.to_csv('/content/drive/Shareddrives/Parrot/Jinho/rock가사중간저장8901-15000.csv', encoding='utf-8')
    print(str(count)+'개 저장완료')
    df_lyric=pd.DataFrame(columns=['가사'])
```
