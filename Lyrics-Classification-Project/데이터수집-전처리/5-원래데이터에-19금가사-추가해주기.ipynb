# indi

```python
indi1 = pd.read_csv('indi최종.csv', encoding='utf-8', index_col=0)
indi1.head(3)
```

```python
indi1.shape
```

```python
indi1.loc[indi1['가사'] == '가사 확인 불가']
```

```python
indi2 = pd.read_csv('indi19금가사추가1-15000.csv', encoding='utf-8', index_col=0)
indi2.head(3)
```

```python
indi2.shape
```

```python
for i in range(len(indi2)):
    if not indi1.loc[indi1['곡 번호'] == indi2['곡 번호'][i], '가사'].empty:
        indi1.loc[indi1['곡 번호'] == indi2['곡 번호'][i], '가사'] = indi2['가사'][i]
```

```python
indi1.loc[indi1['가사'] == '가사 확인 불가']
```

```python
indi1.tocsv('indi.csv',encoding='utf-8')
```
