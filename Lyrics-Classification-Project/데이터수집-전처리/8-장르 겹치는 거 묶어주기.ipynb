```python
def integrate_Genre(g):
  return ', '.join(sorted(g.unique()))
a = df.groupby('가사').agg({'제목':'first',
                      '가수':'first',
                      '곡 번호':'first',
                      '장르': integrate_Genre}).reset_index()
a
```

```python
sorted(a['장르'].unique())
```

```python
a[a.duplicated(['제목','가수','가사'])] 
```

```python
a[a.duplicated(['가사'])] 
```

```python
a = a[['제목', '가수', '곡 번호', '가사', '장르']]
a
```

```python
a.to_csv('/content/drive/Shareddrives/Parrot/lyrics/97694곡-중복제거.csv',encoding='utf-8')
```

```python
a['발라드'] = np.where(a['장르'].str.contains('ballad'), 1, 0)
a['댄스'] = np.where(a['장르'].str.contains('dance'), 1, 0)
a['힙합'] = np.where(a['장르'].str.contains('hiphop'), 1, 0)
a['R&B'] = np.where(a['장르'].str.contains('R&B/Soul'), 1, 0)
a['인디'] = np.where(a['장르'].str.contains('indi'), 1, 0)
a['락'] = np.where(a['장르'].str.contains('rock'), 1, 0)
a['트로트'] = np.where(a['장르'].str.contains('trot'), 1, 0)
a['포크'] = np.where(a['장르'].str.contains('folk'), 1, 0)

```
